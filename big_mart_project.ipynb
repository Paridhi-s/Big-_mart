```python
#catcon
def catcon(df):
  cat=[]
  con=[]
  for i in df.columns:
    if df[i].dtypes=="object":
      cat.append(i)
    else:
      con.append(i)
  return cat,con


def anova_con(df,con_col):
  cat,con=catcon(df)
  from statsmodels.formula.api import ols
  from statsmodels.stats.anova import anova_lm
  imp_cat=[]
  for i in cat:
    reg=con_col+"~"+i
    model=ols(reg,df).fit()
    results=anova_lm(model)
    print("---------------------",con_col,"vs",i,"----------------------------")
    print("pvalue: ",results.iloc[0,4])
    if results.iloc[0,4]<0.05:
      imp_cat.append(i)
  print("\n","Important Categorical features: ",imp_cat)
  return imp_cat
```

```python
from PM6 import preprocessing
Xnew = preprocessing(X)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(Xnew,Y,test_size = 0.2,random_state = 21)

from statsmodels.api import OLS,add_constant
xconst = add_constant(xtrain)
ols = OLS(ytrain,xconst)
model = ols.fit()
model.summary()
```

# **Model 1 (Best Model)**

```python
from warnings import filterwarnings
filterwarnings("ignore")

drop_cols = list(model.pvalues.sort_values().index)[-1]
print("dropped_column:",drop_cols,"\n")
Xnew = Xnew.drop(labels=[drop_cols],axis=1)
xtrain,xtest,ytrain,ytest = train_test_split(Xnew,Y,test_size = 0.2,random_state = 21)

from statsmodels.api import OLS,add_constant
xconst = add_constant(xtrain)
ols = OLS(ytrain,xconst)
model = ols.fit()
print("--------------------------------------------------------")
print("Adjusted Rsquared",round(model.rsquared_adj,5))
print("Rsquared",round(model.rsquared,5))
print("--------------------------------------------------------")
```
